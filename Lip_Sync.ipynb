```python
import cv2


image = cv2.imread("./input/Screenshot.png")


fps = 30  # Frames per second

video_duration = 30  # Video duration in seconds


total_frames = int(fps * video_duration)

video_writer = cv2.VideoWriter("./input/input_video.mp4", cv2.VideoWriter_fourcc('m','p','e','g'), fps, (image.shape[1], image.shape[0]))


for _ in range(total_frames):

    video_writer.write(image)


video_writer.release()

print("Video created successfully!")
```
